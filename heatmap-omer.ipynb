```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

protocols = ['GBN', 'SR', 'SW']

# Plotting heatmaps for each protocol
fig, axes = plt.subplots(1, 3, figsize=(18, 6), sharey=True)
for i, protocol in enumerate(protocols):
    # Read data from CSV
    df = pd.read_csv(f"{protocol}_protocol_times.csv", index_col=0)
    axes[i].set_title(f"{protocol} Protocol Time for Transfer")
    # Plot heatmap
    sns.heatmap(df, annot=True, fmt=".1f", cmap="coolwarm", ax=axes[i])
    axes[i].set_xlabel("Latency")
    axes[i].set_ylabel("Packet Loss Rate")

plt.tight_layout()
plt.show()
```
